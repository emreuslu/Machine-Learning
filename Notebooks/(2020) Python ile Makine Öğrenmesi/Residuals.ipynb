# **Artıklar (Residuals)**

**MSE : Hata Karaler Ortalaması**                         
**RMSE: Hata Kareler Ortalamasının Karekökü**

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
df = pd.read_csv("./Advertising.csv")
df = df.iloc[:,1:len(df)]# index problemini gidermek için
df.head()
X = df[["TV"]]
y = df[["sales"]]
reg = LinearRegression()  #regresyon nesnesi oluşturuldu.
model = reg.fit(X,y)
```

```python
y.head()
```

```python
model.predict(X)   #modelimizle X bağımsız değişkenini kullanarak tahmin yaptık
```

```python
predicted_Y = pd.DataFrame(model.predict(X))
```

```python
predicted_Y[:10]
```

```python
residuals = pd.concat([y,predicted_Y],axis=1)
```

```python
residuals.columns = (["y","Predicted Y"])
```

```python
residuals
```

```python
residuals["Residuals"] = residuals["y"] - residuals["Predicted Y"]
```

```python
residuals
```

```python
residuals["MSE"] = (residuals["Residuals"] ** 2 ).mean()
```

```python
residuals
```

```python
residuals["RMSE"] = np.sqrt((residuals["Residuals"] ** 2 )).mean()
```
